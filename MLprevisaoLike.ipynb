```python
# Previsão de dados ML

import pandas 
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
```

```python
df = pandas.read_csv('curtidas.csv')

d = {'N':0,'D':1}
df['shift'] = df['shift'].map(d)
d = {'Ba':0,'Es':1,'Rj':2}
df['state'] = df['state'].map(d)
d = {'YES':1,'NO':0}
df['like'] = df['like'].map(d)

features = ['Age','time','state','shift']

X = df[features]
y = df['like']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

print(dtree.predict([[31,3,0,1]]))
print("no like[0]")
print("yes like[1]")
```
